```python
import requests
from PIL import Image
from io import BytesIO
import openai
from dotenv import load_dotenv
```

```python
# Load environment variables from .env file
load_dotenv()
```

```python
client = openai.OpenAI()  # Uses your OPENAI_API_KEY from environment
```

```python
response = client.images.generate(
    model="dall-e-3",  # or "dall-e-2" if you don't have access to 3
    prompt="A realistic close-up portrait of an older man with deep wrinkles, sharp blue eyes, and a long white beard, wearing a traditional Viking helmet with intricate designs. The background is a blurred image of a snowy landscape, enhancing the focus",
    n=1,
    size="1024x1024"
)
```

```python

image_url = response.data[0].url
print(image_url)
```

```python
response_img = requests.get(image_url)
img = Image.open(BytesIO(response_img.content))
img.show()  # Opens in default viewer

# For inline display in Jupyter Notebook:
img
# ...existing code...
```
